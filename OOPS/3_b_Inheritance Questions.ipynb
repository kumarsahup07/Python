# Inheritance in Python - Practice Questions

## Question 1: Simple Inheritance

```python
class Animal:
    def sound(self):
        return 'Some sound'

class Dog(Animal):
    def sound(self):
        return 'Bark'

obj = Dog()
print(obj.sound())
```

What will be the output of the code above?

````python
class Animal:
    def sound(self):
        return 'Some sound'

class Dog(Animal):
    def sound(self):
        return 'Bark'

obj = Dog()
print(obj.sound())
````

## Question 4: Hierarchical Inheritance

```python
class Parent:
    def common(self):
        return 'Parent method'

class Child1(Parent):
    def specific(self):
        return 'Child1 method'

class Child2(Parent):
    def specific(self):
        return 'Child2 method'

obj1 = Child1()
obj2 = Child2()
print(obj1.common())
print(obj2.specific())
```

What will be the output of the code above?

````python
class Parent:
    def common(self):
        return 'Parent method'

class Child1(Parent):
    def specific(self):
        return 'Child1 method'

class Child2(Parent):
    def specific(self):
        return 'Child2 method'

obj1 = Child1()
obj2 = Child2()
print(obj1.common())
print(obj2.specific())
````

## Question 10: Overriding and MRO

```python
class A:
    def __init__(self):
        self.value = 'A'

class B(A):
    def __init__(self):
        super().__init__()
        self.value = 'B'

class C(A):
    def __init__(self):
        super().__init__()
        self.value = 'C'

class D(B, C):
    def __init__(self):
        super().__init__()

obj = D()
print(obj.value)
```

What will be the output of the code above?

````python
class A:
    def __init__(self):
        self.value = 'A'

class B(A):
    def __init__(self):
        super().__init__()
        

class C(A):
    def __init__(self):
        super().__init__()
        self.value = 'C'

class D(B, C):
    def __init__(self):
        super().__init__()

obj = D()
print(obj.value)
````

````python
D.mro()
````
