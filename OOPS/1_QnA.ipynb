#### 3. Create a class representing a Rectangle with attributes length and width. Implement methods to calculate its area and perimeter.

```python
class Rectangle:
    def __init__(self,l,b):
        self.length = l
        self.breadth = b
    def area(self):
        print(f'Area of Rectangle is : {self.length*self.breadth}')
    def perimeter(self):
        print(f'Area of Rectangle is : {2*(self.length+self.breadth)}')
        
rect = Rectangle(5,6)
rect.area()
rect.perimeter()
```

#### 4. Define a class for a BankAccount with methods to deposit, withdraw, and check balance. Ensure proper encapsulation of data.

```python
class BankAccount:
    def __init__(self,bal=0) -> None:
        self.balance = bal
    def deposit(self,dep=0):
        self.balance += dep
        print(f"{dep} is deposited into your account")
    def withdraw(self,wd):
        self.balance -= wd
        print(f'{wd} is withdrawn from your account')
    def check_bal(self):
        print(f'Current available balance: {self.balance}')

acc = BankAccount()
acc.deposit(19000)
acc.withdraw(5600)
acc.check_bal()
```

#### 5. Write a Python programme with a class which has attributes like title, actor and actress in its constructor and another method which prints all these properties.
- (a) Prompt the user to enter all this information and add this info in a list.
- (b) Keep the prompt open add all this information until user reply in ’No’ and then display all this information of the list.

```python
class Movie:
    def __init__(self,title,actor,actress) -> None:
        self.title = title
        self.actor = actor
        self.actress = actress
    def show(self):
        print(f'Movie name: {self.title}, Actor Name: {self.actor} and Actress Name: {self.actress}')
    
movies = []

while True:
    title = input("Enter title of the movie")
    actor_name = input("Enter actor_name of the movie")
    actress_name = input("Enter actress_name of the movie")
    
    obj = Movie(title,actor_name,actress_name)
    movies.append(obj)
    choice = input("Enter choice").strip().lower()
    if choice=='no':
        break
    
print('-'*10,'All movies Entered','-'*10)
for m in movies:
    m.show()
```

#### 7. Define a class for a TodoList with methods to add, remove, and display tasks. Ensure tasks are stored as objects of another class

```python
class Task:
    def __init__(self, description, priority, due_date):
        self.description = description
        self.priority = priority
        self.due_date = due_date

    def __str__(self):
        return f"{self.description} | Priority: {self.priority} | Due: {self.due_date}"


class TodoList:
    def __init__(self):
        self.tasks = []   # store Task objects

    def add_task(self, description, priority, due_date):
        task = Task(description, priority, due_date)
        self.tasks.append(task)
        print("Task added successfully!")

    def remove_task(self, description):
        for task in self.tasks:
            if task.description == description:
                self.tasks.remove(task)
                print("Task removed successfully!")
                return
        print("Task not found.")

    def display_tasks(self):
        if not self.tasks:
            print("No tasks available.")
        else:
            print("\nYour Tasks:")
            for i, task in enumerate(self.tasks, start=1):
                print(f"{i}. {task}")

todo = TodoList()

todo.add_task("Complete assignment", "High", "2025-02-15")
todo.add_task("Buy groceries", "Medium", "2025-02-12")

todo.display_tasks()
```

#### 9. Implement a class representing a Bank with methods to add accounts, remove accounts, and display total balance across all accounts.

```python
class Account:
    def __init__(self,acc_number,acc_holder,acc_balance) -> None:
        self.acc_number = acc_number
        self.acc_holder = acc_holder
        self.acc_balance = acc_balance
    def __str__(self):
        return f"Account({self.acc_number}) | Holder: {self.acc_holder} | Balance: {self.acc_balance}"
    

class Bank:
    def __init__(self) -> None:
        self.accounts = list()
        
    def add_account(self,acc_number,acc_holder,acc_balance):
        acc = Account(acc_number,acc_holder,acc_balance)
        self.accounts.append(acc)
        print(f"Account No. {acc_number} added successfully")
        
    def remove_account(self,acc_number):
        for ac in self.accounts:
            if ac.acc_number == acc_number:
                self.accounts.remove(ac)
                print(f"Account no. {acc_number} removed successfully")
                return
        print(f'Account not found')
        
    def total_balance(self):
        total = sum(acc.acc_balance for acc in self.accounts)
        print(f"Total Balance in Bank: {total}")
        
    def display_accounts(self):
        if not self.accounts:
            print("No accounts available.")
        else:
            print("\nBank Accounts:")
            for acc in self.accounts:
                print(acc)
                
                
bank = Bank()

bank.add_account(101, "Alice", 5000)
bank.add_account(102, "Bob", 3000)
bank.add_account(103, "Charlie", 7000)

bank.display_accounts()

bank.remove_account(102)

bank.total_balance()
```
