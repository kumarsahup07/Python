# ## **1ï¸âƒ£ What is Abstraction?**

Abstraction means **hiding unnecessary details** and showing only the essential features.

It allows you to define a **common structure** while forcing subclasses to implement specific behaviors.

Python achieves abstraction using:

### âœ… `ABC` â†’ Abstract Base Class

### âœ… `@abstractmethod` â†’ Marks methods that must be implemented

---

# ## **2ï¸âƒ£ Why Do We Use Abstraction?**

* Ensures **standard structure** for all subclasses
* Forces child classes to implement required methods
* Helps achieve **polymorphism**
* Prevents creating incomplete objects

Example:
Every *User* must **login** and **authenticate**, but each one may do it differently.

---

# ## **3ï¸âƒ£ Creating an Abstract Class in Python**

```python
from abc import ABC, abstractmethod

class User(ABC):

    @abstractmethod
    def login(self):
        pass

    def logout(self):
        print("Logged out")

    @abstractmethod
    def auth(self):
        pass
```

### âœ” You **cannot** create an object of `User`

```python
u = User()   # âŒ Error: Can't instantiate abstract class
```

---

# ## **4ï¸âƒ£ Implementing Subclasses**

### **Buyer Class**

```python
class Buyer(User):

    def login(self):
        if self.auth():
            print("Logging in Buyer")

    def auth(self):
        return True
```

### **Seller Class**

```python
class Seller(User):

    def login(self):
        if self.auth():
            print("Logging in Seller")

    def auth(self):
        print("OTP Verified")
        return True
```

# # ğŸŸ¦ **Real-Life Example: User â†’ Admin, Customer, Seller**

A perfect real-world abstraction scenario.

---

# ## ğŸ”¹ **Abstract Base Class: User**

```python
from abc import ABC, abstractmethod

class User(ABC):

    @abstractmethod
    def login(self):
        pass

    @abstractmethod
    def permissions(self):
        pass
```

---

# ## ğŸ”¹ Customer Class

```python
class Customer(User):

    def login(self):
        print("Customer login via email/password")

    def permissions(self):
        return ["browse_products", "add_to_cart", "place_order"]
```

---

# ## ğŸ”¹ Seller Class

```python
class Seller(User):

    def login(self):
        print("Seller login via OTP verification")

    def permissions(self):
        return ["add_products", "view_sales", "manage_inventory"]
```

---

# ## ğŸ”¹ Admin Class

```python
class Admin(User):

    def login(self):
        print("Admin login via 2FA authentication")

    def permissions(self):
        return ["add_users", "remove_users", "view_reports", "system_settings"]
```

---

# ## âœ” Usage Example

```python
users = [Customer(), Seller(), Admin()]

for u in users:
    u.login()
    print("Permissions:", u.permissions())
    print()
```

### **Output**

```
Customer login via email/password
Permissions: ['browse_products', 'add_to_cart', 'place_order']

Seller login via OTP verification
Permissions: ['add_products', 'view_sales', 'manage_inventory']

Admin login via 2FA authentication
Permissions: ['add_users', 'remove_users', 'view_reports', 'system_settings']
```

---

# ## â­ What This Demonstrates

| Concept              | Explanation                                       |
| -------------------- | ------------------------------------------------- |
| **Abstraction**      | All users must have `login()` and `permissions()` |
| **Polymorphism**     | Same method â†’ different behavior                  |
| **Common Interface** | Every user object can be used in a loop uniformly |
| **Security**         | Enforces structure across the system              |

# # ğŸŸ¦ **Real-Life Banking Abstraction Example**

Abstraction hides **internal details** and exposes **only essential functions** to the user.

In banking:

* Customers don't see **how transactions are processed internally**
* They only see features like: `withdraw()`, `deposit()`, `check_balance()`, etc.

Letâ€™s model this in Python using an **abstract class**.

---

# ## ğŸ”¹ **Abstract Base Class: BankAccount**

```python
from abc import ABC, abstractmethod

class BankAccount(ABC):

    @abstractmethod
    def deposit(self, amount):
        pass

    @abstractmethod
    def withdraw(self, amount):
        pass

    def check_balance(self):
        print("Balance information is confidential.")
```

Here:

* Internal logic of deposit/withdraw is **hidden**
* Only method definitions are exposed

---

# ## ğŸ”¹ Savings Account (Concrete Class)

```python
class SavingsAccount(BankAccount):
    def __init__(self, balance):
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited {amount}. New Balance = {self.balance}")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds")
        else:
            self.balance -= amount
            print(f"Withdrawn {amount}. New Balance = {self.balance}")
```

---

# ## ğŸ”¹ Current Account (Another Subclass)

```python
class CurrentAccount(BankAccount):
    def __init__(self, balance, overdraft_limit):
        self.balance = balance
        self.limit = overdraft_limit

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited {amount}. New Balance = {self.balance}")

    def withdraw(self, amount):
        if amount > self.balance + self.limit:
            print("Overdraft limit exceeded")
        else:
            self.balance -= amount
            print(f"Withdrawn {amount}. New Balance = {self.balance}")
```

---

# ## ğŸ”¹ Using the Abstraction

```python
acc1 = SavingsAccount(5000)
acc2 = CurrentAccount(2000, overdraft_limit=5000)

acc1.deposit(1000)
acc1.withdraw(7000)

acc2.withdraw(6000)
```

---

# ## â­ What This Demonstrates

| Banking Action | Hidden Internal Logic (Abstraction)              |
| -------------- | ------------------------------------------------ |
| Deposit        | Update ledger, verify account, store transaction |
| Withdraw       | Check balance, update database, authentication   |
| Check Balance  | Retrieve data from secure servers                |

User only interacts with simple methods; complexity is hidden.

# # ğŸŸ§ **Abstraction vs Encapsulation (Clear Comparison)**

A very common interview question. Here is the **cleanest** possible explanation:

---

# ## ğŸŸ¦ **Abstraction**

### ğŸ”¹ What it is:

Hides **implementation details** and shows only **essential features**.

### ğŸ”¹ Purpose:

To **simplify complexity**.

### ğŸ”¹ Achieved Using:

* Abstract classes (`ABC`)
* Abstract methods (`@abstractmethod`)
* Interfaces
* Method overriding

### ğŸ”¹ Example:

You call `withdraw(1000)`, but the bank hides:

* Database checks
* Authentication
* Balance verification
* Transaction recording

---

# ## ğŸŸ© **Encapsulation**

### ğŸ”¹ What it is:

Hiding **data** inside the class and controlling access using:

* Public attributes
* Protected attributes
* Private attributes

### ğŸ”¹ Purpose:

To **protect data** and maintain security.

### ğŸ”¹ Achieved Using:

* `_protected_var`
* `__private_var`
* Getters & setters

### ğŸ”¹ Example:

```python
class Bank:
    def __init__(self):
        self.__balance = 5000   # private

    def deposit(self, amount):
        self.__balance += amount

    def check_balance(self):
        return self.__balance
```

User cannot access `__balance` directly â†’ **Encapsulation**.

---
