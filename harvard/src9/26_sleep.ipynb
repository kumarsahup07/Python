```python
# Returns a list of sheep


def main():
    n = int(input("What's n? "))
    for s in sheep(n):
        print(s)


def sheep(n):
    flock = []
    for i in range(n):
        flock.append("ğŸ‘" * i)
    return flock


if __name__ == "__main__":
    main()
```

```python
# Uses yield


def main():
    n = int(input("What's n? "))
    for s in sheep(n):
        print(s)


def sheep(n):
    for i in range(n):
        yield "ğŸ‘" * i


if __name__ == "__main__":
    main()
```
